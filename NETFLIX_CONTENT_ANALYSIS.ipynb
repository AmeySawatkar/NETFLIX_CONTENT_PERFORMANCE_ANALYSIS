```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload file manually
uploaded = files.upload()

# Load dataset
df = pd.read_csv("netflix_movies_detailed_up_to_2025.csv")

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

# Basic info
print("Dataset Shape:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Records:", df.duplicated().sum())
```

```python
# ============================================================
# Step 2: Data Cleaning
# ============================================================

# 1. Drop duration column because it is fully null
df.drop(columns=['duration'], inplace=True)

# 2. Fill missing values in important text columns
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['genres'] = df['genres'].fillna('Unknown')
df['description'] = df['description'].fillna('Unknown')

# 3. Final check after cleaning
print("Columns Removed Successfully!")
print("\nNew Dataset Shape:", df.shape)

print("\nRemaining Missing Values:")
print(df.isnull().sum())
```

```python
# ============================================================
# Step 3: Check Unique Values
# ============================================================

categorical_columns = df.select_dtypes(include='object').columns

for column in categorical_columns:
    print(f"\n========== {column} ==========")
    print(df[column].nunique(), "unique values")
    print(df[column].dropna().unique()[:20])   # first 20 unique values only
```

```python
# ============================================================
# Step 4: Save Cleaned Dataset
# ============================================================

df.to_csv("netflix_content_cleaned.csv", index=False)
print("Cleaned dataset saved successfully as netflix_content_cleaned.csv")
```
